### Feature Engineering

```python
#using this to check the data column we are going to create
ames[['yr_sold','mo_sold']].head()
```

```python
#since we can see that there is some effect of the year on the sale price, we will look deeper into it

#let's create a date_sold column
ames['date_sold']=pd.to_datetime(ames['mo_sold'].astype(str)+'-'+ames['yr_sold'].astype(str))
```

```python
#check against the yr and mo sold printed above
ames['date_sold'].head()
```

```python
#let's see the date_sold to saleprices now
saleprice_yr=ames.groupby('date_sold')['saleprice'].mean()

plt.figure(figsize=(12,10))
plt.plot(saleprice_yr)

plt.title('Mean Sale Prices of Homes Sold by date',fontsize=25)
plt.ylabel('Sale Prices of Homes Sold ($)',fontsize=20)
plt.xlabel('Date Sold',fontsize=20,);
```

```python
plt.figure(figsize=(12,10))
plt.scatter(ames['year_built'],ames['saleprice'])

plt.title('Mean Sale Prices of Homes Sold by Year Built',fontsize=25)
plt.ylabel('Sale Prices of Homes Sold ($)',fontsize=20)
plt.xlabel('Year Built',fontsize=20,);
```

```python
ames['saleprice'].corr(ames['year_built'])
```

```python
#seems like there is a positive relationship between year_built and saleprcies
#this means that the newer the house the higher the price
#but let's see this clearer in an age variable
ames['age']=ames['yr_sold']-ames['year_built']
```

```python
plt.figure(figsize=(12,10))
plt.scatter(ames['age'],ames['saleprice'])

plt.title('Mean Sale Prices of Homes Sold by Age',fontsize=25)
plt.ylabel('Sale Prices of Homes Sold ($)',fontsize=20)
plt.xlabel('Age',fontsize=20,);
```

```python
ames['saleprice'].corr(ames['age'])
```

```python
#another variable that will affect price is alos the number of stories
#so we seek to make house_style a dummy variable of two_story
# 1 for 2-stories and 0 for 1 story

#first,we create a dict for the dummies
two_stories_dict={'1Story':0,'1.5Fin':0,'1.5Unf':0,
                  '2Story':1,'2.5Fin':1,'2.5Unf':1,
                  'SFoyer':1, 'SLvl':1}

ames['two_stories']=ames['house_style'].map(two_stories_dict)
```

```python
#we create a dummy variable for if the hosue has a garage
ames['garage']=ames['garage_type'].map(lambda x: 0 if x=='0' else 1)
```

```python
##we create a dummy variable for if the hosue has central aircon
ames['c_aircon']=ames['central_air'].map(lambda x: 1 if x=='Y' else 0)
```

```python
kitchen_qual_dict={'Ex':1,'Gd':2,'TA':3,
                  'Fa':4,'Po':5}

ames['kitchen_qual_nom']=ames['kitchen_qual'].map(kitchen_qual_dict)
```

```python
ames['artery_dum']=ames['condition_1'].map(lambda x: 1 if x=='Artery' else 0)
```

```python
ames['ns_rail_dum']=ames['condition_1'].map(lambda x: 1 if x=='RRAn' else 0)
```

```python
ames['ew_rail_dum']=ames['condition_1'].map(lambda x: 1 if x=='RRAe' else 0)
```

```python
ames['feedr']=ames['condition_1'].map(lambda x: 1 if x=='Feedr' else 0)
```

```python
#another variable that will affect price is alos the neighborhood
#since this is an indicator of location
#so we seek to make dummy variables for neighborhood 

neighborhood_dums=pd.get_dummies(ames['neighborhood'],
                                 prefix='nbhd',drop_first=True)
#concat to ames df
ames=pd.concat([ames,neighborhood_dums],axis=1)
```
